```python
import pandas as pd
import matplotlib.pyplot as plt

# Четене на данни
file_path_realistic = "/content/realistic_sales_data.csv"
df = pd.read_csv(file_path_realistic)

# 1: Зареждане на първите записи
# Четене на първите 5 записа
data_preview = df.head()
print("Preview of the data:\n", data_preview)

# Проверка за липсващи стойности
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

# 2: Изчисляване на ключови показатели
# 2.1 Средна цена на продуктите
average_price = df["Price"].mean().round(2)
print("Average Price:", average_price)

# 2.2 Обща стойност на продажбите
df["Total Sales Value"] = df["Sales"] * df["Price"]
total_sales_value = df["Total Sales Value"].sum()
print("Total Sales Value:", total_sales_value.round(2))

# 2.3 Среден брой продажби по регион и продукт
average_sales_by_region = df.groupby("Region")["Sales"].mean().round(2).reset_index()
print("Average Sales by Region:\n", average_sales_by_region)

average_sales_by_product = df.groupby("Product")["Sales"].mean().round(2).reset_index()
print("Average Sales by Product:\n", average_sales_by_product)

# 2.4 Продукт и регион с най-високи продажби
top_product = df.groupby("Product")["Sales"].sum().idxmax()
top_region = df.groupby("Region")["Sales"].sum().idxmax()
print("Top-Selling Product:", top_product)
print("Top-Selling Region:", top_region)

# Task 3: Визуализация

# 3.1 Стълбчеста диаграма: Общи продажби по региони
region_sales = df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(8, 6))
region_sales.plot(kind="bar", title="Total Sales by Region", xlabel="Region", ylabel="Sales")
plt.show()

# 3.3 Кръгова диаграма: Дел на продажбите на продуктите спрямо общите продажби
product_sales = df.groupby("Product")["Sales"].sum()
plt.figure(figsize=(8, 8))
product_sales.plot(kind="pie", title="Sales Share by Product", autopct="%1.1f%%")
plt.ylabel("")  # Премахване на y-label
plt.show()

# 3.4 Scatter plot: Цена спрямо обем продажби
plt.figure(figsize=(8, 6))
plt.scatter(df["Price"], df["Sales"], alpha=0.5)
plt.title("Relationship Between Price and Sales")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.show()
```
